```python
import itk
import pyvista as pv

from physiomotion4d import ContourTools


template_model_surface = pv.read("average_surface.vtp")

contour_tools = ContourTools()

reference_image = contour_tools.create_reference_image(
    template_model_surface,
)

template_mask = contour_tools.create_mask_from_mesh(
    template_model_surface,
    reference_image,
)

itk.imwrite(template_mask, "average_binary_mask.nii.gz")


# Then use 3D Slicer to assign the following labels:
# 1: heart muscle
# 2: right ventricle
# 3: left ventricle
# 4: right atrium
# 5: left atrium
# 6: background rim around heart muscle

# save the 3D Slicer labelmap file as "average_labelmap.nii.gz"
```
