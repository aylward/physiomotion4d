```python
import os
import itk

from physiomotion4d.transform_tools import TransformTools
```

```python
os.makedirs("results_CombineModels", exist_ok=True)
```

```python
all_mask_t00 = itk.imread("results/Case1Pack_T00_mask_org.mha")
lung_mask_t30 = itk.imread("results/Case1Pack_T30_lung_mask_org.mha")
other_mask_t30 = itk.imread("results/Case1Pack_T30_other_mask_org.mha")

img_tfm_lung_t00 = itk.transformread("results/Case1Pack_T00_dynamic_forward.hdf")
img_tfm_other_t00 = itk.transformread("results/Case1Pack_T00_static_forward.hdf")
```

```python
TfmTools = TransformTools()

lung_mask_t00 = TfmTools.transform_image(lung_mask_t30, img_tfm_lung_t00, all_mask_t00)
other_mask_t00 = TfmTools.transform_image(other_mask_t30, img_tfm_other_t00, all_mask_t00)

itk.imwrite(lung_mask_t00, "results_CombineModels/lung_mask_t00.mha", compression=True)
itk.imwrite(other_mask_t00, "results_CombineModels/other_mask_t00.mha", compression=True)
```

```python
import pyvista as pv
```

```python
lung_model_t30 = pv.read("TransformedModels/DirLab-4DCT/Case1Pack_T30_lung_lungGatedBase.vtp")
other_model_t30 = pv.read("TransformedModels/DirLab-4DCT/Case1Pack_T30_other_lungGatedBase.vtp")

model_tfm_lung_t00 = itk.transformread(
    "TransformedImages/DirLab-4DCT/Case1Pack_T00_dynamic_inverse.hdf"
)
model_tfm_other_t00 = itk.transformread(
    "TransformedImages/DirLab-4DCT/Case1Pack_T00_static_inverse.hdf"
)

lung_model_t00 = TfmTools.transform_pvcontour(lung_model_t30, model_tfm_lung_t00)
other_model_t00 = TfmTools.transform_pvcontour(other_model_t30, model_tfm_other_t00)

lung_model_t00.save("Experiment_CombineModels/lung_model_t00.vtp")
other_model_t00.save("Experiment_CombineModels/other_model_t00.vtp")
```
