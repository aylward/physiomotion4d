```python
from pathlib import Path

import os

from physiomotion4d.workflow_convert_vtk_to_usd import WorkflowConvertVTKToUSD
```

```python
vtknames = ["a", "la", "lca", "lv", "myo", "pa", "ra", "rv"]
usdnames = [
    "Aorta",
    "LeftAtrium",
    "LeftCoronaryArtery",
    "LeftVentricle",
    "Myocardium",
    "PulmonaryArtery",
    "RightAtrium",
    "RightVentricle",
]

for vtkname, usdname in zip(vtknames, usdnames):
    if os.path.exists(Path.absolute(Path(f"RVOT28-Dias-{usdname}.usd"))):
        os.remove(Path.absolute(Path(f"RVOT28-Dias-{usdname}.usd")))
    converter = WorkflowConvertVTKToUSD(
        vtk_files=[f"../../data/CHOP-Valve4D/Simpleware/parts/{vtkname}.vtk"],
        output_usd=Path.absolute(Path(f"RVOT28-Dias-{usdname}.usd")),
        separate_by_connectivity=False,
        separate_by_cell_type=False,
        mesh_name=f"RVOT28Dias_{usdname}",
        appearance="anatomy",
        anatomy_type="heart",
    )
    converter.run()
```
